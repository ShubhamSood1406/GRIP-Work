# Decision tree for Iris data classification

```
# Importing libraries in Python
import sklearn.datasets as datasets
import pandas as pd

# Loading the iris dataset
iris=datasets.load_iris()

# Forming the iris dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))

y=iris.target
print(y)
```

```

# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(criterion = 'entropy', max_depth=6)
dtree.fit(df,y)

print('Decision Tree Classifer Created')
```

```
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

# To predict in which class the new values goes into

```
decision_class = (dtree.predict([[6.9,0.2,11.2,1.3]])[0])

if decision_class == 0:
  d_c = "Setosa"
elif decision_class == 1:
  d_c = "Verginica"
else:
  d_c = "Versicolor"


print('Resultant class : {}'.format(d_c)) # 0 - setosa, 1 - verginica , 2 - versicolor
```
